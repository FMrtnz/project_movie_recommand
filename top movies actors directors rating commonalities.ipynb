```python


titlesbasics = pd.read_csv('title.basics.tsv.gz',usecols=['primaryTitle', 'genres', 'tconst'], sep = '\t')
```

```python
titlesbasics
```

```python
titlesbasics["genres"] = titlesbasics["genres"].str.split(",")

titlesbasics = titlesbasics.explode("genres")
titlesbasics
```

```python
titleRatings = pd.read_csv('title.ratings.tsv.gz', sep = '\t')
titleRatingsSrtd = titleRatings.sort_values( by=['numVotes', 'averageRating'], ascending= False )
titleRatingsSrtd
```

```python
titleRatingsSrtd['POP SCORE'] = titleRatingsSrtd['averageRating'] * titleRatingsSrtd['numVotes']
```

```python
titleRatingsSrtd
```

```python
titlesBasicsRatings = titlesbasics.merge(titleRatingsSrtd, how= 'left', on= 'tconst')
```

```python
titleSRTDpop = titlesBasicsRatings.sort_values( by=['POP SCORE'], ascending= False )
```
